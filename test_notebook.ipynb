```python
from app import app
from app import db
from app.models import Census
```

```python
import csv
census_data = csv.DictReader(open("data/CENSUS_STATEAGESEX.csv"))
for record in census_data:
    census_row = Census(state = record['STATE'],
                        sex = record['SEX'],
                        age = record['AGE'],
                        pop2000 = record['POP2000'],
                        pop2008 = record['POP2008'])
    print "{0}, {1}, {2}, {3}".format(census_row.state, census_row.sex, census_row.age, census_row.pop2000
                                      , census_row.pop2008)
    db.session.add(census_row)
```
